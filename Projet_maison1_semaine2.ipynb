- Ecrivez une fonction qui détermine la commune (CP Ville) la plus proche d'un point à partir de sa latitude et sa longitude.
- Ajoutez une fonction de conversion pour pouvoir utiliser la première fonction avec un GPS (degrés, minutes, secondes).


```python
# import des modules usuels
import numpy as np
import pandas as pd
import math  as math
# options d'affichage
pd.set_option("display.min_rows", 16)
```

```python
# chargement des données
geo = pd.read_csv("correspondance-code-insee-code-postal.csv",
                   sep=';',
                   usecols=range(11))
geo
```

```python
#Ajoutez une colonne 'CP Ville' avec le Code postal + un espace + et le nom de la Ville.
geo["CP Ville"]=geo.apply(lambda x : x["Code Postal"] + " " + x["Commune"], axis = 1)
geo.head()
```

```python
# on calcule  la latitude et la longitude et on ajoute les colones
geo["Latitude"] = geo["geo_point_2d"].apply(lambda x: float(x.split(', ')[0]))
geo["Longitude"] = geo["geo_point_2d"].apply(lambda x: float(x.split(', ')[1]))
geo.head()
```

```python
#Ecrivez une fonction qui détermine la commune (CP Ville) la plus proche d'un point à partir de sa latitude et sa longitude.

#distance entre deux points sur la terre en sachant latitude et longitude
def distance(A, B):
    R = 6371
    pi = 3.1415927
    lat_alpha = pi * A[0]/180 
    lat_beta = pi * B[0]/180 
    long_alpha = pi * A[1]/180
    long_beta = pi * B[1]/180
    phi = abs(long_alpha - long_beta)
    p = math.acos(math.sin(lat_beta)* math.sin(lat_alpha)+math.cos(lat_beta)* math.cos(lat_alpha)*math.cos(phi))
    d = p * R
    return d
    
# fonction recherche de ville

def ville(lat, long):
    #I put lat and long into a list
    l= [lat, long]
    #I create a column with all the distances
    geo["Distances"] = geo.apply(lambda x: distance([x["Latitude"], x["Longitude"]], l), axis =1)
    #compute the minimum of the distances
    m = geo["Distances"].min()
    #now I have to see to which CP Ville corresponds
    result = geo.loc[geo["Distances"] == m]
    return result["CP Ville"]
```

```python
# on applique la fonction à une coordonnée tirée au hasard
np.random.seed(0)
a, b = 41.5, 51.1  # latitude min et max de la France métropolitaine
lat = (b - a) * np.random.random() + a
a, b = -5.1, 9.5  # longitude min et max de la France métropolitaine
long = (b - a) * np.random.random() + a

ville(lat, long)
```

```python
# conversion degrés, minutes, secondes => décimal
def dms2dec(deg, mn, sec):
    dec = deg + mn/60 + sec /3600
    return dec
```

```python
# à partir de coordonnées GPS précises
ville(dms2dec(48, 42, 52), dms2dec(2, 14, 45))
```
